## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['gross_sales'] = df['unit_price'] * df['qty']
df2=df.loc[:,['unit_price','qty','gross_sales']]
df2.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def ship_price(weight, qty):
    net_weight = weight * qty
    if net_weight < 50:
        return net_weight * 10
    else:
        return net_weight * 7

df['ship_price'] = df.apply(lambda x: ship_price(x['unit_weight'],x['qty']), axis = 1)
df3=df.loc[:,['unit_price',	'unit_weight',	'qty',	'ship_price']]
df3.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%



df['total_price'] = (df['gross_sales']+df['ship_price']) * 1.0925
df4=df.loc[:,['gross_sales','ship_price','total_price']]
df4.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

def cost(qty, unit_cost):
    cogs = qty * unit_cost
    return cogs


df['total_cost'] = df.apply(lambda x: cost(x['qty'], x['unit_cost']), 
                            axis=1) + df['ship_price']


df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['profit'] = df['total_price'] - df['total_cost']
df.head()
```

```python


 
ts=df[df['client_id'].isin(top_5)].groupby('client_id')['qty'].sum() #finds the total quantity of units purchased for each of the the top 5 customers
ship=df[df['client_id'].isin(top_5)].groupby('client_id')['ship_price'].sum() #finds the total shipping costs inclurreds for each of the top 5 customers
profit=df[df['client_id'].isin(top_5)].groupby('client_id')['profit'].sum()#finds the total profit for each of the top 5 customers
rev=df[df['client_id'].isin(top_5)].groupby(['client_id'])['total_price'].sum()#finds the total profit for each of the top 5 customers

#Extract the index column into a list that will be changed into the first column of a dataframe below.
client5_id_list = df_ID.index.tolist()


# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.


top_5_df =pd.DataFrame(data=client5_id_list, columns=['client_id'])
'''I need to use either a join or merge, using the client ID as the common bond.'''
# Merged the various df into one df.
top_5_df=pd.merge(top_5_df,ts, on='client_id')
top_5_df=pd.merge(top_5_df,ship, on='client_id')
top_5_df=pd.merge(top_5_df,rev, on='client_id')
top_5_df=pd.merge(top_5_df,profit, on='client_id')
#top_5_percent_df=top_5_df['profit']/top_5_df['total_price']
top_5_df


```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.


def format_comma (x):
    return '{:,.0f}'.format(x)

def format_dollar(x):
    return '${:,.2f}M'.format(x)

def convert_to_int(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return pd.NA  # Convert non-convertible values to NA

# Changes column titles
top_5_df=top_5_df.rename(columns={"client_id":"Client ID", 'qty':'Units Shipped',"ship_price":'Shipping','total_price':'Total Revenues','profit':'Total Profit'})

# Changes values to millions

top_5_df[['Shipping', 'Total Revenues', 'Total Profit']] = \
    top_5_df[['Shipping', 'Total Revenues', 'Total Profit']]/1000000

# Format $M
top_5_df['Units Shipped'] = top_5_df['Units Shipped'].apply(format_comma)
top_5_df[['Shipping', 'Total Revenues', 'Total Profit']] = \
    top_5_df[['Shipping', 'Total Revenues', 'Total Profit']].applymap(format_dollar)


top_5_df
```

```python
# Sort the updated data by "Total Profit" from highest to lowest
top_5_df= top_5_df.sort_values(["Total Profit"], ascending=False)
top_5_df
```
